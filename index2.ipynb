```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./auto-mpg.data",delim_whitespace=True,header=None)
df.head()
```

```python
new_columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
df.columns=new_columns
df.head()
```

```python
df.isna().sum()
```

```python
df =df.dropna()
```

```python
df_dummies = pd.get_dummies(df,columns=['origin'],prefix='origin')
df_dummies.head()
```

```python
df= pd.get_dummies(df,columns=['origin'],drop_first=True,dtype=int)
df.head()
```

```python
y=df['mpg']
x1=df[['weight']]
x1=sm.add_constant(x1)
model=sm.OLS(y,x1).fit()
model.summary()
```

```python
y=df['mpg']
X=df[['weight','model year','origin_2.0','origin_3.0']]
X=sm.add_constant(X)

model = sm.OLS(y,X).fit()
model.summary()
```
